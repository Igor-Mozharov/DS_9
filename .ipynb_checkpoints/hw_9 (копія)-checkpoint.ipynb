```python
import numpy as np
from tensorflow import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense
from keras import losses, metrics, optimizers
from keras.utils import to_categorical
from keras.layers.regularization import dropout
```

```python
num_classes = 10
num_features = 784
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
x_train, x_test = np.array(x_train, np.float32), np.array(x_test, np.float32)
x_train, x_test = np.reshape(x_train, (-1, num_features)), np.reshape(x_test, (-1, num_features))
x_train, x_test = x_train / 255., x_test / 255.
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
model = Sequential()
model.add(Dense(800, activation='relu', input_dim=num_features))
model.add(Dense(num_classes, activation='softmax'))
model.compile(optimizer=optimizers.SGD(learning_rate=0.01),
              metrics=[metrics.accuracy],
              loss=losses.categorical_crossentropy)
model.summary()
```

```python
result = model.fit(x_train, y_train, batch_size=1000, epochs=500, validation_split=0.2)
```
